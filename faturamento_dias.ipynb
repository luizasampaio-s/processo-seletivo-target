```python
import pandas as pd

df = pd.read_json('dados.json')
```

```python
df
```

```python
df = df.query('valor > 0')
df
```

```python
df_min_filter = df['valor'].min()

df_min = df.query('valor == @df_min_filter')
df_min
```

```python
df_max_filter = df['valor'].max()
df_max = df.query('valor == @df_max_filter')
df_max
```

```python
df_media = df['valor'].mean()

taller_than_average = df.query(" valor > @df_media")

taller_than_average['dia'].count()
```
